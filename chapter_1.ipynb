```python
# Import necessary libraries
# - matplotlib: For creating visualizations
# - numpy: For numerical computations
# - pandas: For handling datasets
# - sklearn.linear_model: For applying Linear Regression models
import matplotlib as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Step 1: Download and prepare the data
# This dataset contains GDP per capita and life satisfaction scores.
# We are fetching it from an online source.
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
```

```python
# Extract the independent (X) and dependent (y) variables
# X: GDP per capita (features)
# y: Life satisfaction (target variable)
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat["Life satisfaction"].values
```

```python
# Step 2: Import additional visualization library for ease of use
import matplotlib.pyplot as plt
```

```python
# Step 3: Visualize the data
# Scatter plot to observe the relationship between GDP and life satisfaction
lifesat.plot(
    kind='scatter', grid=True, x="GDP per capita (USD)", y="Life satisfaction"
)
```

```python
# Define axis limits for better visualization
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
# Step 4: Initialize and use a Linear Regression model
# Linear Regression helps to find the best-fit line predicting y from X
model = LinearRegression()
```

```python
# Fit the model using the data
# The `fit` method calculates the optimal line through the data points.
model.fit(X, y)
```

```python
# Step 5: Predict life satisfaction for a specific GDP value
# Example: Predict for a GDP of $40,000
predicted_lifesat = model.predict([[40_000]])
```

```python
# Print the predicted life satisfaction
print(f"Predicted life satisfaction for $40,000 GDP per capita: {predicted_lifesat[0]:.2f}")
```
